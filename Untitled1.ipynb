```python
import pandas as pd
import pyximport
import numpy as np
pyximport.install(setup_args={"include_dirs": np.get_include()},
                  reload_support=True)
from algorithms.knn_neighborhood import UserKNN
from surprise import Reader, Dataset
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv("data/ml-100k/ml-100k/u.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"], usecols=["user_id", "item_id", "rating"])
data_df["user_id"] = data_df["user_id"].map({b: a for a, b in enumerate(data_df["user_id"].unique())})
data_df["item_id"] = data_df["item_id"].map({b: a for a, b in enumerate(data_df["item_id"].unique())})

reader = Reader(rating_scale=(1, 5))
train_df, test_df = train_test_split(data_df, test_size=0.2)
dataset = Dataset.load_from_df(data_df, reader=reader)
raw_trainset = [(ruid, riid, r, None) for ruid, riid, r in train_df.to_records(index=False)]
raw_testset = [(ruid, riid, r, None) for ruid, riid, r in test_df.to_records(index=False)]
trainset = Dataset.construct_trainset(dataset, raw_trainset)
testset = Dataset.construct_testset(dataset, raw_testset)
```

```python
sim = UserKNN.compute_similarities(trainset, min_support=1)
pop = UserKNN.compute_popularities(trainset)
gain = UserKNN.compute_gain(trainset)
```

```python
def f(tau_pop=0, tau_gain=0, k=1):
    N = set()
    for iuid in trainset.all_users():
        
        simranks = {v: k for k, v in dict(enumerate(np.argsort(sim[iuid, :]))).items()}
        popranks = {v: k for k, v in dict(enumerate(np.argsort(pop))).items()}
        gainranks = {v: k for k, v in dict(enumerate(np.argsort(gain[iuid, :]))).items()}
        overall_ranks = {}
        for candidate in trainset.all_users():
            if candidate != iuid:
                rank = (1-tau_pop-tau_gain) * simranks[candidate] + tau_pop * popranks[candidate] + tau_gain * gainranks[candidate]
                overall_ranks[candidate] = rank
        topk_neighbors = [iuid for iuid, _ in sorted(overall_ranks.items(), key=lambda t: t[1])[::-1][:k]]
        N = N.union(topk_neighbors)
    return len(N)
```

```python
N_sim = []
N_pop = []
N_gain = []
Ks = [5, 10, 15, 30]
for k in Ks:
    print(k)
    # sim only
    N_sim.append(f(k=k))
    
    # pop only
    N_pop.append(f(tau_pop=1, k=k))
    
    # gain_only
    N_gain.append(f(tau_gain=1, k=k))

plt.plot(Ks, np.array(N_sim) / trainset.n_users, label="Sim only")
plt.plot(Ks, np.array(N_pop) / trainset.n_users, label="Pop only")
plt.plot(Ks, np.array(N_gain) / trainset.n_users, label="Gain only")
plt.legend()
plt.ylabel("Frac. of unique neighbors")
plt.xlabel("k")
```
