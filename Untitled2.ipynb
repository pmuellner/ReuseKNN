```python
#Ks = [5, 10, 15, 20, 25, 30]
Ks = [10]
results = load_data(name="ml-100k", protected=True)

def test_significance(c, item_popularity):
    labels = ["UserKNN", "UserKNN+Reuse", "Popularity", "Popularity+Reuse", "Gain", "Gain+Reuse", "DP"]
    n_items = len(item_popularity)
    for method in labels:
        vs = []
        for k in range(len(Ks)):
            r1 = c["UserKNN"][k]
            z1 = 0.5 * (np.log(1 + r1) - np.log(1 - r1))

            r2 = c[method][k]
            z2 = 0.5 * (np.log(1 + r2) - np.log(1 - r2))

            N1 = N2 = n_items
            v = (z1 - z2) / np.sqrt((1 / (N1 - 3)) + (1 / (N2 - 3)))
            vs.append(v)
        for k_idx in range(len(Ks)):
            print("[%s] %f/%f" % (method, c[method][k_idx], vs[k_idx]))
        print()

def plot_k_r(ax, correlations): 
    ax.plot(Ks, correlations["DP"], color="gray", linestyle="dashed", alpha=0.75)
    ax.plot(Ks, correlations["UserKNN"], color="C0", linestyle="dashed", alpha=0.5)
    ax.plot(Ks, correlations["Popularity"], color="C1", linestyle="dashed", alpha=0.5)
    ax.plot(Ks, correlations["Gain"], color="C2", linestyle="dashed", alpha=0.5)
    ax.plot(Ks, correlations["UserKNN+Reuse"], color="C0", linestyle="solid")
    ax.plot(Ks, correlations["Popularity+Reuse"], color="C1", linestyle="solid")
    ax.plot(Ks, correlations["Gain+Reuse"], color="C2", linestyle="solid")

def get_correlation(name, Ks, item_popularity):
    def pearson(d_k):
        x, y = [], []
        for iid, freq in d_k.items():
            x.append(item_popularity[iid])
            y.append(freq)
        r, p = pearsonr(x, y)
        return r

    corr = defaultdict(list)
    for k_idx, k in enumerate(Ks):
        d = pl.load(open("results/protected/" + name + "/item_frequency_userknn_full.pkl", "rb"))
        corr["DP"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_userknn.pkl", "rb"))
        corr["UserKNN"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_userknn_reuse.pkl", "rb"))
        corr["UserKNN+Reuse"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_pop.pkl", "rb"))
        corr["Popularity"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_pop_reuse.pkl", "rb"))
        corr["Popularity+Reuse"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_gain.pkl", "rb"))
        corr["Gain"].append(pearson(d[k_idx]))
        d = pl.load(open("results/protected/" + name + "/item_frequency_gain_reuse.pkl", "rb"))
        corr["Gain+Reuse"].append(pearson(d[k_idx]))
    
    return corr
    
    
df = pd.read_csv("data/ml-100k/u.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"])
item_popularity = df.groupby("item_id").size()
c = get_correlation("ml-100k", Ks, item_popularity)
#plot_k_r(axes[0], c)
test_significance(c, item_popularity)
```

```python
ranked_sim = np.zeros((trainset.n_users,trainset.n_users))
ranked_gain = np.zeros((trainset.n_users,trainset.n_users))
for target_user in trainset.all_users():
    for rank, neighbor in enumerate(np.argsort(sim[target_user, :])):
        ranked_sim[target_user, neighbor] = rank
    for rank, neighbor in enumerate(np.argsort(gain[target_user, :])):
        ranked_gain[target_user, neighbor] = rank

ranked_pop = np.zeros(trainset.n_users)
for rank, neighbor in enumerate(np.argsort(pop)):
    ranked_pop[neighbor] = rank
```

```python
corr_sim_gain = []
for user in trainset.all_users():
    r, p = spearmanr(ranked_sim[user], ranked_gain[user])
    if p < 0.001:
        corr_sim_gain.append(r)
        
np.mean(corr_sim_gain), len(corr_sim_gain) / trainset.n_users
```

```python
k = 10
selected_neighbors = []
for target_user in trainset.all_users():
    neighbor_rank = ranked_sim[target_user] + ranked_gain[target_user]
    selected_neighbors.append(np.argsort(neighbor_rank)[-k:])
selected_neighbors = np.array(selected_neighbors)

neighbor_sims, neighbor_simranks = [], []
neighbor_gains, neighbor_gainranks = [], []
for target_user in trainset.all_users():
    neighbor_sim = (sim[target_user, selected_neighbors[target_user]] / sim[target_user].max()).mean()
    neighbor_gain = (gain[target_user, selected_neighbors[target_user]] / gain[target_user].max()).mean()
    neighbor_sims.append(neighbor_sim)
    neighbor_gains.append(neighbor_gain)
    
    neighbor_simrank = (ranked_sim[target_user, selected_neighbors[target_user]] / ranked_sim[target_user].max()).mean()
    neighbor_gainrank = (ranked_gain[target_user, selected_neighbors[target_user]] / ranked_gain[target_user].max()).mean()
    neighbor_simranks.append(neighbor_simrank)
    neighbor_gainranks.append(neighbor_gainrank)
```
